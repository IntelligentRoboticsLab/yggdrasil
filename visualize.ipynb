```python
import numpy as np

def calculate_center_of_pressure(pressure_values, sensor_positions):
    """
    Calculate the center of pressure (CoP) given pressure values and sensor positions.

    Args:
        pressure_values (numpy array): Array of pressure values from each sensor.
        sensor_positions (numpy array): Array of (x, y) coordinates of each sensor.

    Returns:
        tuple: A tuple containing the (CoPx, CoPy) coordinates of the center of pressure.
    """
    
    # compute total pressure
    total_pressure = np.sum(pressure_values)
    
    # if robot in the air we can't average
    if total_pressure == 0:
        return 0, 0
    
    # compute center of pressure
    CoPx = np.sum(pressure_values * sensor_positions[:, 0]) / total_pressure
    CoPy = np.sum(pressure_values * sensor_positions[:, 1]) / total_pressure

    return CoPx, CoPy
```

```python
# make relative positions within range [0,1] [0,1]
sensor_positions_left = np.array([(-1,-1), (-1,1), (1,-1), (1,1)])
sensor_positions_right = np.array([(-1,-1), (-1,1), (1,-1), (1,1)])
```

```python
def compute_center(row, foot='left'):
    # get the pressure values in an array
    pressure_left = np.array([row[f'{foot}_foot_fsr_front_left'], row[f'{foot}_foot_fsr_front_right'], row[f'{foot}_foot_fsr_rear_left'], row[f'{foot}_foot_fsr_rear_right']])
    
    # compute the center of pressure
    center = calculate_center_of_pressure(pressure_left, sensor_positions_left)
    
    # return as a tuple
    return (center)

# compute the center of pressure for each foot
df['center_left'] = df.apply(compute_center, axis=1, args=('left',))
df['center_right'] = df.apply(compute_center, axis=1, args=('right',))
```

```python
df['center_left_x'] = df['center_left'].apply(lambda x: x[0])
df['center_left_y'] = df['center_left'].apply(lambda x: x[1])

# for right
df['center_right_x'] = df['center_right'].apply(lambda x: x[0])
df['center_right_y'] = df['center_right'].apply(lambda x: x[1])
```
